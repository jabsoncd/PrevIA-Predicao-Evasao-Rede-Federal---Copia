```python
import matplotlib.pyplot as plt

# Dados dos modelos e suas mdias de accuracy
modelos = [
    "CatBoost categrico", "CatBoost", "XGBoost", "LightGBM",
    "SVM (e Linear SVC)", "Random Forest", "Logistic Regression",
    "KNN", "Decision Tree"
]
accuracy = [0.72, 0.71, 0.71, 0.71, 0.68, 0.67, 0.67, 0.67, 0.65]

# Identificar os 3 melhores
top_indices = sorted(range(len(accuracy)), key=lambda i: accuracy[i], reverse=True)[:3]

# Cores
cores_vivas = ['#e41a1c', '#377eb8', '#4daf4a']  # Vermelho, Azul, Verde
cor_neutra = '#cccccc'

# Criar grfico
plt.figure(figsize=(12, 6))
for i, (modelo, acc) in enumerate(zip(modelos, accuracy)):
    cor = cores_vivas[top_indices.index(i)] if i in top_indices else cor_neutra
    plt.plot([0, 1], [0, acc], label=modelo, color=cor, linewidth=2 if i in top_indices else 1)
    
    # Adiciona valor ao final da linha
    plt.text(1.01, acc, f"{acc:.2f}", va='center', fontsize=9, color=cor)

# Ajustes do grfico
plt.title("Desempenho dos Modelos (Accuracy)  Destaque nos Top 3")
plt.xlabel("Eixo X (0 a 1)")
plt.ylabel("Accuracy")
plt.xlim(0, 1.15)
plt.ylim(0, 1)
plt.legend(loc="lower right", bbox_to_anchor=(1.35, 0))
plt.grid(True)
plt.tight_layout()

# Exibir grfico
plt.show()
```
